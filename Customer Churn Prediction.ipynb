```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/HP/Downloads/cp.csv")
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df = pd.get_dummies(df)
```

```python
df=df.dropna()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error,accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
x=df.drop(['churn'],axis=1)
y=df['churn']
```

```python
from sklearn.model_selection import train_test_split
train_x, valid_x, train_y, valid_y = train_test_split(x, y, test_size = 0.3, random_state = 43,stratify=y)

train_x.shape, valid_x.shape, train_y.shape, valid_y.shape
```

```python
rfc=RandomForestClassifier(n_jobs=-1,n_estimators=70,min_samples_split=185,max_depth=12,criterion='entropy')
rfc.fit(train_x,train_y)
pred_train=rfc.predict(train_x)
pred_test=rfc.predict(valid_x)
print(accuracy_score(valid_y,pred_test))
print(accuracy_score(train_y,pred_train))
```

```python
confusion_matrix(valid_y,pred_test)
```

```python
confusion_matrix(train_y,pred_train)
```
